Q2) Use Jupyter Notebook to execute following code, try to guess what is happening in this code.

Answer: It is a graph plot representing the height distribution of various people.


```python
import numpy as np
import matplotlib.pyplot as plt
height=np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.hist(height, bins=30,ec='gold')
```
